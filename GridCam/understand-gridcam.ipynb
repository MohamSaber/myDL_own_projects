```python
import tensorflow as tf
```

```python
from tensorflow.keras.applications import ResNet50
```

```python
resnet=ResNet50(include_top=True,weights='imagenet')
```

```python
resnet.summary()
```

```python
import numpy as np
img_path = "/kaggle/input/chickadee/chickadee.png"

img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = tf.keras.applications.resnet50.preprocess_input(img_array)
```

```python
preds = resnet.predict(img_array)
print("Predicted:", tf.keras.applications.resnet50.decode_predictions(preds, top=1)[0])
```

```python
def get_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]

    grads = tape.gradient(class_channel, conv_outputs)
    weights = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(weights, conv_outputs), axis=-1)

    heatmap = np.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()
```

```python
last_conv_layer_name = "conv5_block3_out"
```

```python
heatmap = get_gradcam_heatmap(img_array, resnet, last_conv_layer_name)
```

```python
import cv2
import matplotlib.pyplot as plt
img_org = cv2.imread(img_path)
img_org = cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB)

heatmap = cv2.resize(heatmap, (img_org.shape[1], img_org.shape[0]))
heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)

superimposed = heatmap_color * 0.4 + img_org

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img_org)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grad-CAM")
plt.imshow(superimposed.astype(np.uint8))
plt.axis("off")
plt.show()
```
